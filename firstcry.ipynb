```python
boy=[]
for i in url:
    source=requests.get(i)
    soup=BeautifulSoup(source.text)

    data=soup.find_all("div",class_="lblock lft")
    for item in data:
        clean_text = item.text.strip()
        clean_text = re.sub(r'\s+', ' ', clean_text)
        boy.append(clean_text)
```

```python
boy
```

```python
structured_data = []

for entry in boy:
    try:
        # Extract status indicators
        status_match = re.findall(r'(Added to Cart|Shortlisted|Already Viewed|[1-3] Left)', entry)
        status = ' | '.join(status_match)

        # Clean out the status from the string
        cleaned_entry = re.sub(r'(Added to Cart|Shortlisted|Already Viewed|[1-3] Left)', '', entry).strip()

        # Match prices and discount
        price_pattern = re.search(
            r'(\d+\.\d+)\s+(\d+)\s+\((?:upto\s+)?(\d+% Off)\)\s+Club Price:\s+(\d+\.\d+)', cleaned_entry)

        if not price_pattern:
            raise ValueError("Price pattern not found")

        discounted_price = float(price_pattern.group(1))
        original_price = int(price_pattern.group(2))
        discount_percent = price_pattern.group(3)
        club_price = float(price_pattern.group(4))

        # Remove price details from string
        price_text = price_pattern.group(0)
        title_part = cleaned_entry.replace(price_text, '').strip()

        # Extract brand name (first 1-3 words based on casing)
        title_words = title_part.split()
        brand = title_words[0]
        if len(title_words) > 1 and title_words[1][0].isupper():
            brand += ' ' + title_words[1]
        if len(title_words) > 2 and title_words[2][0].isupper():
            brand += ' ' + title_words[2]

        title = title_part[len(brand):].strip()

        structured_data.append({
            'Status': status,
            'Brand': brand.strip(),
            'Title': title,
            'Discounted Price': discounted_price,
            'Original Price': original_price,
            'Discount Percent': discount_percent,
            'Club Price': club_price
        })
    except Exception as e:
        print(f"Error processing entry: {entry}\n{e}")

df = pd.DataFrame(structured_data)
```
